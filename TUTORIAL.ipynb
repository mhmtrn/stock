```python
# son parametreyi değiştirelim. İki yok geriye gidip, oradan itibaren geçmiş 5 yılı tarayacak.
sm.get_stock_ongoing_performance("equity", 12, 5, 2)
```

```python
# get_stock_ongoing_performance fonksiyonunu çoklusu. Birden fazla oran veya hesap belirlemek için!
# mode varsayılan olarak False. Çünkü sadece sağlayanları istiyoruz.
sm.filter_financials(["turnover", "ebitda"], 12, 3, 4)
```

```python
# iki yıl öncesine giderek geriye doğru referans alalım
sm.filter_financials(["turnover", "ebit"], 12, 4, 3)
```

```python
# iki yıl öncesine giderek geriye doğru referans alalım
sm.filter_financials(["turnover", "ebit", "gross_profit_margin"], 12, 3, 5)
```

### ALTMAN SCORE

- The Altman Z-score is a formula for determining whether a company, notably in the manufacturing space, is headed for bankruptcy. 

- The formula takes into account profitability, leverage, liquidity, solvency, and activity ratios. 

- An Altman Z-score close to 1.8 suggests a company might be headed for bankruptcy, while a score closer to 3 suggests a company is in solid financial positioning.

#### ---> Altman Score Calculation - Basic

```python
sm.altman("PGSUS", term="2015")
```

```python
altman = sm.altman("PGSUS", term="2015")
```

```python
altman["AltmanScore"]
```
